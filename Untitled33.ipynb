```python

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image = cv2.imread('다운로드 (1).jpg')
rows, cols = image.shape[0:2] 

dx, dy = 100, 50

mtrx = np.float32([[1, 0, dx],[0, 1, dy]])

dst= cv2.warpAffine(image, mtrx, (cols+dx, rows+dy))

dst2= cv2.warpAffine(image, mtrx, (cols+dx, rows+dy), None, cv2. INTER_LINEAR, cv2.BORDER_CONSTANT, (255,0,0))

dst3 = cv2.warpAffine(image, mtrx, (cols+dx, rows+dy), None, cv2. INTER_LINEAR, cv2.BORDER_REFLECT)

cv2_imshow(image)
cv2_imshow(dst)
cv2_imshow(dst2)
cv2_imshow(dst3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread('다운로드 (1).jpg')
rows, cols = image.shape[0:2]

d45 = 45.0 * np.pi / 180

d90 = 90.0 * np.pi / 180 

m45 = np.float32([[ np.cos (d45), -1* np.sin(d45), rows//2], [np.sin(d45), np.cos(d45), -1*cols//4]]) 
m90 = np.float32([[ np.cos(d90), -1*np.sin(d90), rows], [np.sin(d90), np.cos(d90), 0]])

r45= cv2.warpAffine(image, m45, (cols, rows)) 
r90 = cv2.warpAffine(image, m90, (rows,cols))

 

cv2_imshow(image)
cv2_imshow(r45)
cv2_imshow(r90)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread('다운로드 (1).jpg')

rows, cols = image.shape[0:2]
#ᅳᅳᅳ1 회전을 위한 변환 행렬 구하기
#회전축 : 중앙, 각도:45, 배율:0.5
m45=cv2.getRotationMatrix2D((cols/2, rows/2), 45,0.5)
#회전축: 중앙, 각도: 90, 배율: 1.5
m90= cv2.getRotationMatrix2D((cols/2, rows/2),90,1.5)
#---2 변환 행렬 적용
img45=cv2.warpAffine(image, m45, (cols, rows))
img90= cv2.warpAffine(image, m90, (cols, rows))
#ᅳᅳᅳ3 결과 출력
cv2_imshow(image)
cv2_imshow(img45)
cv2_imshow(img90)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread('다운로드 (1).jpg') 
height, width= image.shape[:2]
#--1 0.5배 축소 변환 행렬
m_small = np.float32([[0.5, 0, 0],
[0, 0.5,0]])
#--2 2배 확대 변환 행렬
m_big = np.float32([[2, 0, 0],[0, 2, 0]])
# --3 보간법 적용 없이 확대 축소
dst1 = cv2.warpAffine(image, m_small, (int(height +0.5), int (width+0.5)))
dst2 = cv2.warpAffine(image, m_big, (int(height+2), int (width+2)))
# --4 보간법 적용한 확대 축소
dst3= cv2.warpAffine(image, m_small, (int(height +0.5), int (width+0.5)), None, cv2. INTER_AREA)
dst4= cv2.warpAffine(image, m_big, (int (height *2), int (width+2)),  None, cv2. INTER_CUBIC)
#결과 출력
cv2_imshow(image)
cv2_imshow(dst1) 



cv2_imshow(dst2)
cv2_imshow(dst3)
cv2_imshow(dst4)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
